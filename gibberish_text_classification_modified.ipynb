# Combining df_sheet3 and df_gibberish to df_main

```python
df_main=pd.concat([df_gibberish,df_sheet3],ignore_index=True)
print(df_main.shape)
print("number of label values are")
print(df_main["Label"].value_counts())
df_main.head()
```

```python
import tensorflow 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocab=20000
tokenizer = Tokenizer(num_words=vocab)
tokenizer.fit_on_texts(df_main["Response"])
```

```python
print("the length of vocabulary size is :",len(tokenizer.word_index))
tokenizer.word_index
```

```python
X = tokenizer.texts_to_sequences(df_main["Response"])
X
```

```python
dummy=0
for i in X:
    if len(i)>dummy:
        dummy=len(i)
max_len=dummy
max_len
```

```python
X = pad_sequences(X, maxlen=max_len, padding="pre")
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df_main["Label"], test_size=25,random_state=0)
```

```python
import keras
```

```python
model = keras.models.Sequential([
    keras.layers.Embedding(vocab, 64),
    keras.layers.GlobalAveragePooling1D(),
    keras.layers.Dense(64, activation="relu"),
    keras.layers.Dropout(rate=0.25),
    keras.layers.Dense(32, activation="relu"),
    keras.layers.Dropout(rate=0.25),
    keras.layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(optimizer="adam",loss='binary_crossentropy',metrics=["acc"])
```

```python
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test),  verbose=1)
```

```python
y_predict=model.predict(X_test)
```

```python
y_predict=y_predict>0.5
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy_score(y_predict,y_test)
```

```python
print(classification_report(y_predict,y_test))
```
